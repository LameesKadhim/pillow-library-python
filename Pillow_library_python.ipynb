## resize and gray scale with pillow



```python

size = (200,200)
saved = "cat2.png"
try:
  image = Image.open('cat.png')
except:
  print('unable to load image')
#resize image
image.thumbnail(size)
image.save(saved)
display(image)

#grey scale
greyscale_image = image.convert('L')
display(greyscale_image)
```

## Filters in pillow library

```python
from PIL import Image,ImageFilter

original = Image.open('ball.jpg')
print('original image')
display(original)
#CONTOUR filter
contour_image = original.filter(ImageFilter.CONTOUR)
print('applying contour filter')
display(contour_image)

#edge enhance filter
edge_enh_image = original.filter(ImageFilter.EDGE_ENHANCE)
print('edge enhance filter')
display(edge_enh_image)

#find edges filter
find_edge_image = original.filter(ImageFilter.FIND_EDGES)
print('find edges filter')
display(find_edge_image)

```

## using subplot to show the images above

```python
import matplotlib.pyplot as plt

# specifying the overall grid size
plt.figure(figsize=(10,10)) 

image_arr = [original, contour_image, edge_enh_image, find_edge_image]
image_text = ['original', 'contour filter', 'edge enhance filter', 'find edge filter']


for i in range(4):
    plt.subplot(2,2,i+1)    # the number of images in the grid is 2*2 (4)
    plt.title(image_text[i]) 
    plt.xticks([])
    plt.yticks([])
    plt.imshow(image_arr[i])
```

## put logo on an image


```python
ball_image = Image.open('ball.jpg')
logo = Image.open('logo.png')
#resize the logo
logo.thumbnail((60,60))
logo.save('logo.png')

#put the logo on the image
ball_image_copy = ball_image.copy()
position = ((ball_image_copy.width - logo.width),(ball_image_copy.height - logo.height))
ball_image_copy.paste(logo,position)

display(ball_image_copy)
```

## rotating an image


```python
from PIL import Image
import matplotlib.pyplot as plt

image = Image.open('cat.png')

#rotating the image
image_rotate_90 = image.rotate(90)

#flip the image
image_flip = image.transpose(Image.FLIP_TOP_BOTTOM)

#transpose image
image_transpose = image.transpose(Image.TRANSPOSE)

plt.figure(figsize=(10,10))
im_arr = [image,image_rotate_90,image_flip,image_transpose]
im_arr_title = ['original image','rotated image 90' ,'flipped image','transposed image']
for i in range (4):
    plt.subplot(2,2,i+1)
    plt.title(im_arr_title[i])
    plt.xticks([])
    plt.yticks([])
    plt.imshow(im_arr[i])
```
